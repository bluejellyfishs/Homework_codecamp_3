```python
df['pcv'] = pd.to_numeric(df['pcv'],errors='coerce')
df['wc'] = pd.to_numeric(df['wc'],errors='coerce')
df['rc'] = pd.to_numeric(df['rc'],errors='coerce')

age_avg = df['age'].mean()
df['age'].fillna(value = age_avg, inplace = True)

pcv_avg = df['pcv'].mean()
df['pcv'].fillna(value = pcv_avg, inplace = True)

wc_avg = df['wc'].mean()
df['wc'].fillna(value = wc_avg, inplace = True)

rc_avg = df['rc'].mean()
df['rc'].fillna(value = rc_avg, inplace = True)

bp_avg = df['bp'].mean()
df['bp'].fillna(value = bp_avg, inplace = True)

sg_avg = df['sg'].mean()
df['sg'].fillna(value = sg_avg, inplace = True)

al_avg = df['al'].mean()
df['al'].fillna(value = al_avg, inplace = True)

su_avg = df['su'].mean()
df['su'].fillna(value = su_avg, inplace = True)

bgr_avg = df['bgr'].mean()
df['su'].fillna(value = bgr_avg, inplace = True)

bu_avg = df['bu'].mean()
df['bu'].fillna(value = bu_avg, inplace = True)

sc_avg = df['sc'].mean()
df['sc'].fillna(value = sc_avg, inplace = True)

sod_avg = df['sod'].mean()
df['sod'].fillna(value = sod_avg, inplace = True)

pot_avg = df['pot'].mean()
df['pot'].fillna(value = pot_avg, inplace = True)

hemo_avg = df['hemo'].mean()
df['hemo'].fillna(value = hemo_avg, inplace = True)

df.dropna(subset=['dm', 'cad', 'appet', 'pe', 'ane', 'pcc', 'ba'], inplace = True)

fig = plt.figure(figsize=(12, 12))
sns.heatmap(df.isnull(), cbar=False, cmap='Accent')
```

```python
pcc = pd.get_dummies(df['pcc'], drop_first=True)
pcc.rename(columns={'present':'present_pcc'}, inplace=True)
pcc
```

```python
df_obj = df.select_dtypes(['object'])
df[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
dm = pd.get_dummies(df['dm'], drop_first=True)
dm.rename(columns={'yes':'yes_dm'}, inplace=True)
dm
```
