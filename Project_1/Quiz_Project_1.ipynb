```python
#2
low = int(input('ใส่ขอบเขตล่าง: '))
high = int(input('ใส่ขอบเขตบน: '))
prime_number = 0
for i in range(low, high+1):
    count = 0;
    for j in range(2, i):
        if (i%j) == 0:
            count += 1;
            break;
    else:
        if i != 1:
            prime_number += 1;
            print('Prime number: {}'.format(i)) 
print('จำนวน prime number ทั้งหมด: {} ตัว'.format(prime_number))  
```
