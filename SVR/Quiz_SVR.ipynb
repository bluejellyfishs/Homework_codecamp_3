```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#12

X = df['sqft_living']
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
#13

X_train = np.array(X_train).reshape(-1, 1)
X_test = np.array(X_test).reshape(-1,1)
y_train = np.array(y_train).reshape(-1, 1)
y_test = np.array(y_test).reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()

X_train = sc_X.fit_transform(X_train)
y_train = sc_y.fit_transform(y_train)
```

```python
#14

from sklearn.svm import SVR

regressor_linear = SVR(kernel = 'linear')
regressor_linear.fit(X_train, y_train.flatten())

regressor_rbf = SVR(kernel = 'rbf')
regressor_rbf.fit(X_train, y_train.flatten())
```

```python
#15

y_predicted_linear = regressor_linear.predict(X_test)

print('MAE: ', metrics.mean_absolute_error(y_test, y_predicted_linear))
print('MSE: ', metrics.mean_squared_error(y_test, y_predicted_linear))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, y_predicted_linear)))
```

```python
y_predicted_rbf = regressor_rbf.predict(X_test)

print('MAE: ', metrics.mean_absolute_error(y_test, y_predicted_rbf))
print('MSE: ', metrics.mean_squared_error(y_test, y_predicted_rbf))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, y_predicted_rbf)))
```

```python
X = df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated']]
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

y_train = np.array(y_train).reshape(-1,1)

sc_X = StandardScaler()
sc_y = StandardScaler()

X_transform = sc_X.fit_transform(X_train)
y_transform = sc_y.fit_transform(y_train)

regressor_linear = SVR(kernel='linear')
regressor_linear.fit(X_transform, y_transform.flatten())

regressor_rbf = SVR(kernel='rbf')
regressor_rbf.fit(X_transform, y_transform.flatten())
```

```python
#16

y_predicted_linear_all = regressor_linear.predict(X_test)

print('MAE: ', metrics.mean_absolute_error(y_test, y_predicted_linear_all))
print('MSE: ', metrics.mean_squared_error(y_test, y_predicted_linear_all))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, y_predicted_linear_all)))
```

```python
y_predicted_rbf_all = regressor_rbf.predict(X_test)

print('MAE: ', metrics.mean_absolute_error(y_test, y_predicted_rbf_all))
print('MSE: ', metrics.mean_squared_error(y_test, y_predicted_rbf_all))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, y_predicted_rbf_all)))
```

```python
#17

X = df['sqft_living']
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)

lm_simple = LinearRegression()
lm_simple.fit(X_train, y_train)
```

```python
X = df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
       'floors', 'waterfront', 'view', 'condition', 'sqft_above',
       'sqft_basement', 'yr_built', 'yr_renovated']]
y = df['price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)

lm_multi = LinearRegression()
lm_multi.fit(X_train, y_train)
```

```python
#18

y_predicted_simple = lm_simple.predict(X_test)

print('MAE: ', metrics.mean_absolute_error(y_test, y_predicted_simple))
print('MSE: ', metrics.mean_squared_error(y_test, y_predicted_simple))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, y_predicted_simple)))
```

```python
y_predicted_multi = lm_multi.predict(X_test)

print('MAE: ', metrics.mean_absolute_error(y_test, y_predicted_multi))
print('MSE: ', metrics.mean_squared_error(y_test, y_predicted_multi))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, y_predicted_multi)))
```
