```python
df2 = df.drop(['result_type'],axis=1)
df2
```

```python
#15 

#Default

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
                                df2.drop(['Type'], axis=1),
                                df2['Type'],
                                test_size = 0.2,
                                random_state = 70)
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
df_predicted = nb.predict(X_test)
df_predicted
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
confusion_matrix(y_test, df_predicted)
```

```python
print('Accuracy = ', accuracy_score(y_test, df_predicted))
print('F1 score = ', f1_score(y_test, df_predicted, average='micro'))
print('Precision = ', precision_score(y_test, df_predicted, average='micro'))
print('Recall = ', recall_score(y_test, df_predicted, average='micro'))
```
