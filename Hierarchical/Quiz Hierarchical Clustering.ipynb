## ข้อ 9

```python
X = df.iloc[:, [0, 4]].values
X
```

```python
import scipy.cluster.hierarchy as sch
```

```python
fig = plt.figure(figsize=(12, 8))
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))

plt.title('Dendrogram')
plt.xlabel('STG')
plt.ylabel('Euclidean distances')
```

## ข้อ 10

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
y_hc
```

## ข้อ 11

```python
x1 = X[y_hc == 0][:,0]
y1 = X[y_hc == 0][:,1]
x2 = X[y_hc == 1][:,0]
y2 = X[y_hc == 1][:,1]
x3 = X[y_hc == 2][:,0]
y3 = X[y_hc == 2][:,1]

fig = plt.figure(figsize=(12, 8))
plt.scatter(x1, y1, s=40, c='blue', label='Cluster1')
plt.scatter(x2, y2, s=40, c='red', label='Cluster2')
plt.scatter(x3, y3, s=40, c='pink', label='Cluster3')
plt.title('Clusters of STG')
plt.xlabel('STG')
plt.ylabel('PEG')
plt.legend()
plt.show()
```
