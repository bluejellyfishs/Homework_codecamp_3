```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
index = ['A', 'B', 'C', 'D']
values = np.random.uniform(50, 500, (4, 4))
columns = ['2014', '2015', '2016', '2017']
df = pd.DataFrame(values, index, columns)
df
```

```python
a_mean = df.loc['A'].mean()
b_mean = df.loc['B'].mean()
c_mean = df.loc['C'].mean()
d_mean = df.loc['D'].mean()

sns.barplot(x=df.index, y=[a_mean, b_mean, c_mean, d_mean])
```

```python
sns.barplot(x=[a_mean, b_mean, c_mean, d_mean], y=df.index)
```

```python
sns.scatterplot(x=df.index, y=df['2017'], data=df)
```

```python
sns.scatterplot(x=df['2017'], y=df.index, data=df)
```

```python
new_df = pd.crosstab(df['Pclass'], df['Sex'])
sns.heatmap(new_df)
```
